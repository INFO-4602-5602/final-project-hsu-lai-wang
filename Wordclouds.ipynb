```python
import os
import matplotlib.pyplot as plt
from collections import defaultdict
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS) 
```

```python
# get reviews
reviews = defaultdict(list)
ratings = []

data_dir = '/data/movie_reviews/aclImdb/train/'
for f in os.listdir(os.path.join(data_dir, 'neg')):
    rating = int(f.split('_')[1].split('.')[0])
    text = open(os.path.join(data_dir, 'neg', f)).read()
    reviews[rating].append(text)
    ratings.append(rating)
for f in os.listdir(os.path.join(data_dir, 'pos')):
    rating = int(f.split('_')[1].split('.')[0])
    text = open(os.path.join(data_dir, 'pos', f)).read()
    reviews[rating].append(text)
    ratings.append(rating)
    
data_dir = '/data/movie_reviews/aclImdb/test/'
for f in os.listdir(os.path.join(data_dir, 'neg')):
    rating = int(f.split('_')[1].split('.')[0])
    text = open(os.path.join(data_dir, 'neg', f)).read()
    reviews[rating].append(text)
    ratings.append(rating)
for f in os.listdir(os.path.join(data_dir, 'pos')):
    rating = int(f.split('_')[1].split('.')[0])
    text = open(os.path.join(data_dir, 'pos', f)).read()
    reviews[rating].append(text)
    ratings.append(rating)

tmp = []
tmp.extend(reviews[1])
tmp.extend(reviews[2])
reviews_1 = " ".join(tmp)

tmp = []
tmp.extend(reviews[3])
tmp.extend(reviews[4])
reviews_2 = " ".join(tmp)

tmp = []
tmp.extend(reviews[5])
tmp.extend(reviews[6])
reviews_3 = " ".join(tmp)

tmp = []
tmp.extend(reviews[7])
tmp.extend(reviews[8])
reviews_4 = " ".join(tmp)

tmp = []
tmp.extend(reviews[9])
tmp.extend(reviews[10])
reviews_5 = " ".join(tmp)

set(ratings)
```

```python
#assert len(reviews_1) + len(reviews_2) + len(reviews_3) + len(reviews_4) + len(reviews_5) == 50000
len(reviews_1), len(reviews_2), len(reviews_3), len(reviews_4), len(reviews_5)
```

```python
# label 1-2
num = 1
wordcloud = WordCloud(width=800, 
                      height=800,
                      background_color ='white',
                      stopwords = stopwords,
                      min_font_size = 10).generate(reviews_1) 
  
# plot the WordCloud image                        
plt.figure(figsize = (10, 10), facecolor=None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.title("Word cloud of ratings 1 & 2")
plt.savefig("figs/wordcloud_{}.png".format(num), bbox_inches='tight')
plt.show() 
```
